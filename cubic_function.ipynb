# Cubic function

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import matplotlib.pyplot as plt
```

$f(x) = x^3$

```python
def cubic(x):
    return x**3
```

$f(x) = 3x^2$

```python
def cubic_derivative(x):
    return 3 * x**2
```

```python
x = np.arange(- 10, 10, 0.01)
```

```python
plt.plot(x, cubic(x), label='Cubic')
plt.plot(x, cubic_derivative(x), label='Derivative')
plt.grid()
plt.legend()
```
