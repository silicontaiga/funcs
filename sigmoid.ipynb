# Sigmoid Function

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
x = np.arange(- 10, 10, 0.01)
```

```python
plt.plot(x, sigmoid(x))
plt.plot(x, sigmoid_derivative(x))
plt.grid()
```
