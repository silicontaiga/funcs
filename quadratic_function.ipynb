# Quadratic function

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import matplotlib.pyplot as plt
```

```python
def quadratic(x):
    return x**2
```

```python
def quadratic_derivative(x):
    return 2 * x
```

```python
x = np.arange(- 10, 10, 0.01)
```

```python
plt.plot(x, quadratic(x))
plt.plot(x, quadratic_derivative(x))
plt.grid()
```
